```
class Rectangle:
    
    def _init_(self,height,width,corner_x,corner_y):
        
        self.height = height
        self.width = width
        self.corner_x = corner_x
        self.corner_y = corner_y
        
    def centre(self):
        
        return print("center at(",self.corner_x,",",self.corner_y,")")
    

    def area(self):
        
        return self.height * self.width
    
    def perimeter(self):
        
        return 2 * (self.height + self.width)


h = int(input("Enter height of rectangle: "))
w = int(input("Enter width of rectangle: "))
cx= int(input("Enter the corner x of rectangle: "))
cy= int(input("Enter the corner y of rectangle: "))
object = Rectangle(h)
object = Rectangle(w)
object = Rectangle(cx)
object = Rectangle(cy)
print(object.centre())
print("Area of rectangle: ", object.area())
print("Perimeter of rectangle: ", object.perimeter())
```
